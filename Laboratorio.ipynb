# Optimización 1

Definir las variables y el objetivo
Sea:
- \( R \): radio de la base del cono.
- \( H \): altura del cono.
- \( r \): radio de la base del cilindro.
- \( h \): altura del cilindro.
Queremos maximizar el volumen \( V \) del cilindro. La fórmula para el volumen de un cilindro es:
$$[
V = \pi r^2 h
]$$

Relación entre \( r \) y \( h \)
Dado que el cilindro está inscrito en el cono, podemos usar la semejanza de triángulos. Observa que:
$$[
\frac{r}{R} = \frac{H - h}{H}
]$$
Despejamos \( r \) en términos de \( h \):
$$[
r = R \frac{H - h}{H}
]$$

Expresar \( V \) en términos de \( h \)
Sustituimos esta expresión de \( r \) en la fórmula de \( V \):
$$[
V = \pi \left(R \frac{H - h}{H}\right)^2 h
]$$
Simplificando, obtenemos:
$$[
V = \pi \frac{R^2 (H - h)^2}{H^2} h
]$$

Derivada para encontrar puntos críticos
Para maximizar \( V \), derivamos con respecto a \( h \):
$$[
V(h) = \pi \frac{R^2}{H^2} (H - h)^2 h
]$$
Calculamos \( V'(h) \) aplicando la regla del producto:
$$[
V'(h) = \pi \frac{R^2}{H^2} \left[ 2(H - h)(-1) h + (H - h)^2 \right]
]$$

Igualamos \( V'(h) = 0 \) para encontrar los puntos críticos.

Evaluar los valores críticos
El valor de \( h \) que maximiza el volumen se puede obtener resolviendo la ecuación \( V'(h) = 0 \). Sustituimos este valor crítico de $$( h )$$ en $$( r = R \frac{H - h}{H} )$$ para obtener el radio \( r \) del cilindro.

Así, encontramos las dimensiones \( r \) y \( h \) que maximizan el volumen del cilindro inscrito en el cono.

           

# Optimización 2

La velocidad \( v \) que minimiza la energía empleada por un pez que nada río arriba en una corriente que tiene velocidad $$( -v_c )$$ donde:

$$( v ):$$ velocidad del pez con respecto al agua.
$$ ( -v_c ):$$ velocidad de la corriente del río, siendo negativa ya que está en dirección opuesta a la del pez.


La energía empleada en recorrer una distancia \( d \) a contracorriente es directamente proporcional al tiempo \( t \) necesario y al cubo de la velocidad efectiva.



La velocidad del pez respecto a la orilla, considerando la corriente en contra, es:
$$[
v_{\text{efectiva}} = v + v_c
]$$



El tiempo \( t \) que el pez necesita para recorrer una distancia \( d \) con velocidad efectiva $$( v + v_c )$$ es:
$$[
t = \frac{d}{v + v_c}
]$$



La energía \( E \) es proporcional al tiempo y al cubo de la velocidad efectiva $$( (v + v_c)^3 )$$
Así que podemos escribir:
$$[
E = k \cdot t \cdot (v + v_c)^3
]$$
donde \( k \) es una constante de proporcionalidad.

Sustituimos $$( t = \frac{d}{v + v_c} )$$ en la expresión de \( E \):
$$[
E = k \cdot \frac{d}{v + v_c} \cdot (v + v_c)^3
]$$
Simplificando, obtenemos:
$$[
E = k \cdot d \cdot (v + v_c)^2
]$$



Queremos encontrar la velocidad \( v \) que minimiza \( E \). Para hacerlo, derivamos \( E \) con respecto a \( v \) y luego igualamos la derivada a cero.

1. Derivamos \( E \) respecto a \( v \):
   $$[
   \frac{dE}{dv} = 2k \cdot d \cdot (v + v_c)
   ]$$

2. Igualamos la derivada a cero para encontrar el punto crítico:
   $$[
   2k \cdot d \cdot (v + v_c) = 0
   ]$$
   
3. Resolviendo esta ecuación, encontramos el valor de \( v \) que minimiza la energía empleada.

Queremos encontrar la longitud de la varilla más larga \( L \) que puede transportarse alrededor de una esquina en un pasillo de 6 pies de ancho, suponiendo que la varilla no puede doblarse.



Imagina el pasillo en forma de "L", donde cada tramo tiene 6 pies de ancho. Supón que la varilla toca la esquina interior y los bordes de ambos pasillos.

Define:
- \( w = 6 \) pies, el ancho del pasillo.
- \( L \), la longitud de la varilla que queremos maximizar.



Para simplificar, coloca el pasillo en un plano de coordenadas donde el punto de la esquina sea el origen. Si la varilla toca ambos bordes y la esquina, entonces forma un triángulo rectángulo donde la varilla es la hipotenusa.

Dividimos la varilla en dos segmentos que tocan los lados del pasillo. Supón que la varilla hace un ángulo $$( \theta )$$ con uno de los pasillos, formando dos segmentos de la varilla que se proyectan en cada tramo del pasillo.



Para la configuración en forma de "L", tenemos la relación:
$$[
L = w \sec \theta + w \csc \theta
]$$
donde $$( \sec \theta = \frac{1}{\cos \theta} )$$ y $$( \csc \theta = \frac{1}{\sin \theta} )$$



Deriva \( L \) con respecto a$$( \theta )$$ luego iguala a cero

Derivada de \( L \):
   $$[
   \frac{dL}{d\theta} = w \sec \theta \tan \theta - w \csc \theta \cot \theta
   ]$$

Igualamos $$( \frac{dL}{d\theta} = 0 )$$ y resolvemos para $$( \theta )$$ para encontrar el ángulo que minimiza \( L \).



Usa el valor de $$( \theta )$$ encontrado para calcular \( L \), obteniendo así la longitud de la varilla más larga que puede transportarse alrededor de la esquina del pasillo.

Esta longitud \( L \) es la solución al problema.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton

def f(x):
    return x**5 - 6 * x**3 + 2 * x - 2

def f_prime(x):
    return 5 * x**4 - 18 * x**2 + 2

x_vals = np.linspace(-3, 3, 1000)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label='f(x) = x^5 - 6x^3 + 2x - 2')
plt.axhline(0, color='gray', linestyle='--')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfica de f(x)')
plt.grid()
plt.legend()
plt.show()

conjeturas_iniciales = [-2, 0, 2]
tolerancia = 0.0001

soluciones = []
for conjetura in conjeturas_iniciales:
    try:
        raiz = newton(f, conjetura, f_prime, tol=tolerancia)
        soluciones.append(raiz)
    except RuntimeError:
        print(f"El método de Newton no convergió para la conjetura inicial {conjetura}")

print("Soluciones encontradas usando el método de Newton:")
for solucion in soluciones:
    print(f"x ≈ {solucion:.5f}")

```

# Punto 2 Newton

```python
import numpy as np

def f(x):
    return x**3 - 2

def f_prime(x):
    return 3 * x**2


def newton_method(x0, tol=0.00005, max_iter=1000):
    x = x0
    for _ in range(max_iter):
        x_next = x - f(x) / f_prime(x)
        if abs(x_next - x) < tol:
            return x_next
        x = x_next
    return x


x0 = 1.0
resultado = newton_method(x0)

print(f"Aproximación de ∛2: {resultado:.5f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# Definimos las ecuacinoes
def f(x, a, b, c, d):
    return a * x**3 + b * x**2 + c * x + d

def f_prime(x, a, b, c):
    return 3 * a * x**2 + 2 * b * x + c

# Sistema de ecuaciones ajustado con las condiciones
def equations(vars):
    a, b, c, d = vars
    return [
        f(2, a, b, c, d) + 3,      # f(2) = -3 --> f(2) + 3 = 0
        f(6, a, b, c, d) - 1,      # f(6) = 1  --> f(6) - 1 = 0
        f_prime(2, a, b, c),       # f'(2) = 0 --> f'(2) = 0
        d
    ]

a, b, c, d = fsolve(equations, (0, 0, 0, 0))

x_vals = np.linspace(0, 8, 400)
y_vals = f(x_vals, a, b, c, d)

plt.plot(x_vals, y_vals, label=r'$f(x) = ax^3 + bx^2 + cx + d$')
plt.scatter([2, 6], [-3, 1], color='red')  # Puntos (2, -3) y (6, 1)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Bosquejo de la función que cumple las condiciones dadas')
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.grid(True)
plt.legend()
plt.show()

print(f"Coeficientes de la función cúbica: a = {a:.5f}, b = {b:.5f}, c = {c:.5f}, d = {d:.5f}")
```

# Exercise 1.1



**Dominio**: La función $( f(x) = \frac{x}{x+1} )$ tiene una restricción en el denominador $( x + 1 = 0 )$. Por lo tanto, el dominio es $( x \neq -1 )$.

**Rango**: Observamos que a medida que $( x \to \infty )$, $( f(x) \to 1 )$, y cuando $( x \to -\infty )$, $( f(x) \to 1 )$ también. Debido a la discontinuidad en $( x = -1 )$, la función no alcanza valores en $( y = 1 )$. Por lo tanto, el rango es $( (-\infty, 1) \cup (1, \infty) )$.

**Simetría**: La función $( f(x) = \frac{x}{x+1} )$ no es par ,no cumple $( f(-x) = f(x) $) ni impar no cumple $( f(-x) = -f(x) )$. Por lo tanto, no tiene simetría respecto al eje $( y )$ ni al origen.

**Intersecciones**:
   - **Intersección con el eje $( y )$**: Para encontrar la intersección con el eje $( y $), sustituimos $( x = 0 )$:

     $$[
     f(0) = \frac{0}{0+1} = 0
     ]$$
     Entonces, la intersección con el eje $( y )$ es el punto $( (0, 0) )$.
   - **Intersección con el eje $( x )$**: Para la intersección con el eje $( x)$, igualamos $( f(x) = 0 )$:
   
     $$[
     \frac{x}{x+1} = 0 \Rightarrow x = 0
     ]$$

     Por lo tanto, la intersección con el eje \( x \) también es \( (0, 0) \).

**Primera derivada**: Calculamos la primera derivada para encontrar puntos críticos y el comportamiento de crecimiento y decrecimiento.
   $$[
   f(x) = \frac{x}{x+1}
   ]$$
   Usamos la regla del cociente:
   $$[
   f'(x) = \frac{(x+1)(1) - x(1)}{(x+1)^2} = \frac{x + 1 - x}{(x+1)^2} = \frac{1}{(x+1)^2}
   ]$$
   Como $( f'(x) = \frac{1}{(x+1)^2} > 0 )$ para $( x \neq -1 )$, la función es creciente en todo su dominio.

**Máximos y mínimos**: Dado que la función es siempre creciente en su dominio, no hay máximos ni mínimos locales.

**Segunda derivada**: Calculamos la segunda derivada para determinar la concavidad y los puntos de inflexión.
   $$[
   f''(x) = -\frac{2}{(x+1)^3}
   ]$$
   Observamos que $( f''(x) < 0 )$ para $( x \neq -1 )$, por lo que la función es cóncava hacia abajo en todo su dominio.

**Asíntotas**:
   - **Asíntota vertical**: La función tiene una asíntota vertical en $( x = -1 )$ porque el denominador se anula en ese punto.
   - **Asíntota horizontal**: A medida que $( x \to \infty )$ o $( x \to -\infty )$, la función tiende a 1. Por lo tanto, $( y = 1 )$ es una asíntota horizontal.



```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, limit, oo
from sympy import Rational


x = symbols('x')
f = x / (x + 1)





f_prime = diff(f, x)
print("Primera derivada de f(x):")
print(f_prime)




f = diff(f_prime, x)
print("\nSegunda derivada de f(x):")
print(f)


vertical = limit(f, x, -1)
horizontal = limit(f, x, oo)

print("\nAsíntota vertical en x = -1:")
print(vertical)

print("\nAsíntota horizontal en y = 1:")
print(horizontal)


x_vals = np.linspace(-5, 5, 400)
y_vals = x_vals / (x_vals + 1)

plt.figure(figsize=(10, 6))


plt.plot(x_vals, y_vals, label=r'$f(x) = \frac{x}{x+1}$', color='blue')

plt.axvline(x=-1, color='red', linestyle='--', label='Asíntota vertical $x=-1$')





plt.axhline(y=1, color='green', linestyle='--', label='asintota horizontal $y=1$')



plt.plot(0, 0, 'go', label='Intersección en (0,0)')


plt.ylim(-10, 10)


plt.xlim(-5, 5)
plt.xlabel('x')
plt.ylabel('f(x)')


plt.title('Gráfica de f(x) = x / (x + 1)')
plt.legend()
plt.grid(True)
plt.show()
```

   ]$$

**Simetría**: La función no es par ni impar, ya que para $( x < 0 )$, $( g(x) = 0 )$, mientras que para $( x > 0 )$, $( g(x) )$ toma valores distintos de cero.

**Intersecciones**:
   - **Intersección con el eje $( y )$**: Para $( x = 0 )$, tenemos:

     $$[

   - **Intersección con el eje $( x )$**: Para la intersección con el eje $( x )$, igualamos $( g(x) = 0 )$. Dado que $( g(x) = 0 )$ para $( x \leq 0 )$, no hay otras intersecciones con el eje $( x )$.

**Primera derivada**: Calculamos la primera derivada para encontrar puntos críticos y el comportamiento de crecimiento y decrecimiento.

   $$[
   g(x) = 3x^2 + 2x

   $$[

**Dominio**: La función $( f(x) = \frac{x^2}{x^2 - 1} )$ tiene una restricción en el denominador $( x^2 - 1 = 0 )$. Resolviendo esta ecuación, tenemos:
   $$[
   x^2 = 1 \Rightarrow x = \pm 1
   ]$$
   Por lo tanto, el dominio de la función es $( x \neq \pm 1 )$, o en notación de intervalo:
   $$[
   (-\infty, -1) \cup (-1, 1) \cup (1, \infty)
   ]$$

**Rango**: Observamos que el valor de $( f(x) )$ tiende a infinito cerca de las discontinuidades en $( x = \pm 1 )$. A medida que $( x \to \infty )$ o $( x \to -\infty )$, la función se aproxima a 1, pero no alcanza todos los valores entre $( (-\infty, -1) \cup (1, \infty) )$.

**Simetría**: La función $( f(x) = \frac{x^2}{x^2 - 1} )$ es una función par, ya que $( f(-x) = f(x) )$. Esto indica que es simétrica respecto al eje $( y )$.

**Intersecciones**:
   - **Intersección con el eje $( y )$**: Para encontrar la intersección con el eje $( y )$, sustituimos $( x = 0 )$:

     $$[
     f(0) = \frac{0^2}{0^2 - 1} = 0
     ]$$

     Por lo tanto, la intersección con el eje $( y )$ es el punto $( (0, 0) )$.
   - **Intersección con el eje $( x )$**: Para la intersección con el eje $( x )$, igualamos $( f(x) = 0 )$:

     $$[
     \frac{x^2}{x^2 - 1} = 0 \Rightarrow x = 0
     ]$$

     Así, la intersección con el eje $( x )$ también es $( (0, 0) )$.

**Primera derivada**: Calculamos la primera derivada para encontrar puntos críticos y el comportamiento de crecimiento y decrecimiento.

   $$[
   f(x) = \frac{x^2}{x^2 - 1}
   ]$$

   Usamos la regla del cociente:

   $$[
   f'(x) = \frac{(x^2 - 1) \cdot 2x - x^2 \cdot 2x}{(x^2 - 1)^2} = \frac{2x}{(x^2 - 1)^2}
   ]$$


**Puntos críticos**: Igualamos $( f'(x) = 0 )$ para encontrar los puntos críticos:

   $$[
   \frac{2x}{(x^2 - 1)^2} = 0 \Rightarrow x = 0
   ]$$
   El único punto crítico es $( x = 0 )$.

**Segunda derivada**: Calculamos la segunda derivada para analizar la concavidad.

   $$[
   f''(x) = \frac{2(x^2 - 1)^2 - 2x \cdot 2x \cdot 2(x^2 - 1)}{(x^2 - 1)^4} = \frac{2(x^2 + 1)}{(x^2 - 1)^3}
   ]$$

   La concavidad cambia en función de los valores de $( x )$, lo que nos permite determinar los puntos de inflexión.

**Asíntotas**:
   - **Asíntota vertical**: La función tiene asíntotas verticales en $( x = \pm 1 )$, ya que el denominador se anula en esos puntos.
   - **Asíntota horizontal**: A medida que $( x \to \infty )$ o $( x \to -\infty )$, la función se aproxima a 1, por lo que $( y = 1 )$ es una asíntota horizontal.



```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 / (x**2 - 1)

x_left = np.linspace(-5, -1.01, 200)


x_right = np.linspace(1.01, 5, 200)

plt.figure(figsize=(10, 6))
plt.plot(x_left, f(x_left), label=r'$f(x) = \frac{x^2}{x^2 - 1}$', color='blue')

plt.plot(x_right, f(x_right), color='blue')


plt.plot(0, 0, 'go', label='Intersección en (0,0)')
plt.axvline(x=-1, color='red', linestyle='--', label='Asíntota vertical $x = -1$')




plt.axvline(x=1, color='red', linestyle='--', label='Asíntota vertical $x = 1$')
plt.axhline(y=1, color='green', linestyle='--', label='Asíntota horizontal $y = 1$')

plt.xlabel('x')





plt.ylabel('f(x)')


plt.title('Gráfica de $f(x) = \\frac{x^2}{x^2 - 1}$')
plt.legend()


plt.ylim(-10, 10)
plt.grid(True)


plt.show()
```
